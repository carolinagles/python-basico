```python
num_1 = input("Primer número a comparar")
num_2 = input("Segundo número a comparar")
num_3 = input("Tercer número a comparar")

if (num_1>=num_2) and (num_1>=num_3):
    print(num_1+ " es el mayor")
else:
    if(num_2>=num_1) and (num_2>=num_3): 
        print(num_2+ " es el mayor") 
    else:
        if(num_3>=num_1) and (num_3>=num_2):
            print(num_3+ " es el mayor")        
```

 ### 5. Pedir un numero por teclado hasta introducir un numero entre 1000 y 1500

```python
num = int(input("Número entre 1000 y 1500"))
while num < 1000 or num > 1500: 
    print("Número fuera del rango, debe ser entre 1000 y 1500")
    num = int(input("Número entre 1000 y 1500"))
    print("Número correcto", num)
```

### 7. Leer 2 números por teclado y mostrar el menú:   
**1. Sumar**  
**2. Restar**  
**3. Multiplicar**  
**4. Dividir**  
**Después, leer por teclado un número entero entre 1 y 4.  
Según la opción elegida, mostrar el resultado de la operación con los dos números introducidos.**

```python
num_1 = int(input("Primer número:"))
num_2 = int(input("Segundo número:"))

print("Menú \n1. Sumar \n2. Restar \n3. Mupliticar \n4. Dividir")

opcion = int(input("Operación seleccioanda:"))

if opcion==1:
    print(f"{num_1} mas {num_2} es {num_1 + num_2}")
if opcion==2:
    print(f"{num_1} menos {num_2} es {num_1 - num_2}")
if opcion==3:
    print(f"{num_1} por {num_2} es {num_1 * num_2}")
if opcion==4:
    print(f"{num_1} entre {num_2} es {num_1 / num_2}")
```
