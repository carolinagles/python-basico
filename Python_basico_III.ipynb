### 2. Calculadora con funciones
**Mediante funciones realiza un programa que simule una calculadora:**
1. **Sumar**  
2. **Restar**  
3. **Multiplicar**

```python
# Calculadora básica
# Fuciones
def suma(valor_1, valor2):
    return valor_1 + valor_2 
def resta(valor_1, valor2):
    return valor_1 - valor_2 
def multiplicacion(valor_1, valor2):
    return valor_1 * valor_2 
def division(valor_1, valor2):
    return valor_1 / valor_2
def Menu():
    print("Menú \n1. Sumar \n2. Restar \n3. Mupliticar \n4. Dividir \nElige opción:"))
def valida(min, max):
    opcion=0
    while opcion<min or opcion>max:
        opcion=int(input("Escribe una opcion entre 1 y 4: "))
    return opcion 

# Programa
valor_1 = int(input("valor 1: "))
valor_2 = int(input("valor 2: "))
Menu()
opcion=valida(1,4)
if opcion==1:
    print(suma(valor_1,valor_2)
if opcion==2:
    print(resta(valor_1,valor_2)
if opcion==3:
    print(multiplicacion(valor_1,valor_2)
if opcion==4:
    print(division(valor_1,valor_2)
```

## Avanzado
### Juego del ahorcado.

```python
palabra=input("Palabra a adivinar: ")
intentos=len(palabra)
print("Tienes " +str(intentos), "intentos") 

ganado=False
aciertos=0
letras_introducidas=[]

while not ganado and intentos > 0:
    letra = input("Introduce una letra: ")
    posicion = 0

    for i in palabra:
        posicion += 1
        if letra == i:
            if letra not in letras_introducidas:
                aciertos += 1
                print("Esa letra existe en la posición", posicion)
        else:
            print("Esa letra no existe en la posición", posicion)

    if aciertos == len(palabra):
        ganado = True

    intentos -= 1

    if letra not in letras_introducidas:
        letras_introducidas.append(letra)

if ganado:
    print("¡Has ganado en", len(palabra) - intentos, "intentos!")
else:
    print("Comienza otra vez...")
```
